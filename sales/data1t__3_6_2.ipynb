```python
import pandas as pd
import numpy as np
import pyspark
import random
import datetime
```

```python
pyspark.__version__
```

```python
from pyspark.sql import SparkSession
# from pyspark.sql.functions import col, mean as mean_, countDistinct, count as count_, max as max_
```

```python
spark = (SparkSession
    .builder
    .master('local[*]')
    .appName('leshakot')
    .getOrCreate()
    )
```

```python
# Создать DataFrame с полями: Дата, UserID, Продукт, Количество, Цена.
np.random.seed(42)
# количество строк
count = 1000

product= ["молоко", "кефир", "хлеб", "курица", "свинина", "говядина", "яйцо", "помидор", "лук"]
year = 2024

id = list(range(1, count + 1))
products = [random.choice(product) for _ in range(count)]
price = [random.randint(100, 5000) for _ in range(count)]
quantities = [random.randint(1, 10) for _ in range(count)]
dates = [datetime.date(year, random.randint(1, 12), random.randint(1, 28)) for _ in range(count)]

columns = ['date', 'user_id', 'product', 'quantities', 'price']
df = spark.createDataFrame(list(zip(dates, id, products, quantities, price)), columns)

df.limit(5).toPandas()
```

```python
# Сохранить сгенерированный DataFrame в формате CSV для последующего анализа.
df.toPandas().to_csv('sales.csv', header=True, index=False)
```
